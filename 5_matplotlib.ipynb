```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_arr = np.arange(10)
x_arr
```

```python
y_arr = x_arr

# визуализация
plt.plot(x_arr, y_arr)

plt.show()
```

```python
# совмещение нескольких графиков в одном окне
plt.figure(figsize=(10,8))

y_arr_1 = x_arr ** 2
y_arr_2 = x_arr ** 3

plt.plot(x_arr, y_arr_1)
plt.plot(x_arr, y_arr_2)

plt.show()
```

```python
# 1-мерный массив из 100 значений
# линейной последовательности в диапазоне от 10 до 10
x_arr_1 = np.linspace(-10, 10, 100)
x_arr_1
```

```python
# функция синуса
y_1 = np.sin(x_arr_1)

# функция гиперболического тангенса
y_2 = np.tanh(x_arr_1)

# функция сигмоиды
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_3 = sigmoid(x_arr_1)

plt.figure(figsize=(10, 8))

plt.plot(x_arr_1, y_1)
plt.plot(x_arr_1, y_2)
plt.plot(x_arr_1, y_3)

plt.show()
```

```python
# цвета линий

y_1 = np.sin(x_arr_1)
y_2 = np.sin(x_arr_1 + 1)
y_3 = np.sin(x_arr_1 + 2)
y_4 = np.sin(x_arr_1 + 3)

# по названию цвета 
plt.plot(x_arr_1, y_1, color="red")
# по первой букве названия цвета из формата RGB или CMYK
plt.plot(x_arr_1, y_2, c="m")
# по 16-ричной кодировке RGB или RGBA
plt.plot(x_arr_1, y_3, c="#097aa2")
# по формату RGB или RGBA в виде кортежа со значениями от 0 до 1
plt.plot(x_arr_1, y_4, c=(0.0, 0.9, 0.15, 1.0))

plt.show()
```

```python
# стили линий
y_1 = np.sin(x_arr_1)
y_2 = np.sin(x_arr_1 + 1)
y_3 = np.sin(x_arr_1 + 2)
y_4 = np.sin(x_arr_1 + 3)

# сплошная линия
# plt.plot(x_arr_1, y_1, c="red", linestyle="solid")
# штриховая линия
plt.plot(x_arr_1, y_1, c="red", ls="dashed")
# пунктирная
plt.plot(x_arr_1, y_2, ls="dotted")
# штрих пунктирная
plt.plot(x_arr_1, y_3, ls="dashdot")


plt.show()
```

```python
# заголовок графика, подписи осей, легенды
# размер окна
plt.figure(figsize=(10, 7))

plt.plot(x_arr_1, np.sin(x_arr_1))
plt.plot(x_arr_1, np.cos(x_arr_1))

# Заголовок
plt.title("Графики синуса и косинуса")

# подписи осей
plt.xlabel("Ось X")
plt.ylabel("Ось Y")

# легенды графика
plt.legend(["синусоида", "косинус"])

plt.show()
```
